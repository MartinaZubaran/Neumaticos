```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
import random
import datetime

urls= ( 'https://neumen.ar/categoria-producto/neumaticos/',
        'https://neumen.ar/categoria-producto/neumaticos/page/2/')
#listas
Neumaticos= []
Marca= []
Precio= []

#funcion para obtener la informacion
for url in urls:
    response = requests.get(url, headers={"User-Agent": 'Mozilla/5.0'})
    response.raise_for_status()
    if response.status_code != 200:
        break
        print("No se pudo cargar lo solicitado")
    soup= BeautifulSoup(response.content, 'html.parser')


    # #buscar lo solicitado
    info_neumatico= [item.get_text() for item in soup.findAll(name= 'h2', attrs= {'class': "woocommerce-loop-product__title"})]
    info_tipo= [item.get_text() for item in soup.findAll(name= 'span', attrs= {'class': "loop-product-categories"})]
    info_precio_n= [item.get_text() for item in soup.findAll(name= 'span', attrs= {'class': "woocommerce-Price-amount amount"})]

    #primer_precios = info_precio_n [::2]  # Suponiendo que el precio de lista es el primer precio

    precios= info_precio_n[1:]
    time.sleep(5)


 # #agregar la info a los
    Neumaticos.extend(info_neumatico)
    Marca.extend(info_tipo)
    Precio.extend(precios)


# Crear un DataFrame de pandas para exportar a Excel
df_neumaticos= pd.DataFrame ({      'Neumaticos': Neumaticos,
      'Marca': Marca,
      'Precio': Precio})

  # Agregar la columna "fecha_scrapping" con la fecha actual
df_neumaticos["fecha_scrapping"] = datetime.date.today()
#listas
Neumaticos= []
Marca= []
Precio= []

# #funcion para obtener la informacion
# for url in urls:
#     response = requests.get(url, headers={"User-Agent": 'Mozilla/5.0'})
#     response.raise_for_status()
#     if response.status_code != 200:
#         print("No se pudo cargar lo solicitado")
#         break
#     soup= BeautifulSoup(response.content, 'html.parser')


#     # #buscar lo solicitado
#     info_neumatico= [item.get_text() for item in soup.findAll(name= 'h2', attrs= {'class': "woocommerce-loop-product__title"})]
#     info_tipo= [item.get_text() for item in soup.findAll(name= 'span', attrs= {'class': "loop-product-categories"})]
#     info_precio_n= [item.get_text() for item in soup.findAll(name= 'span', attrs= {'class': "woocommerce-Price-amount amount"})]

#     #primer_precios = info_precio_n [::2]  # Suponiendo que el precio de lista es el primer precio

#     precios= info_precio_n[1:]
#     time.sleep(5)


#  # #agregar la info a los
#     Neumaticos.extend(info_neumatico)
#     Marca.extend(info_tipo)
#     Precio.extend(precios)


# # Crear un DataFrame de pandas para exportar a Excel
# df_neumaticos= pd.DataFrame ({      'Neumaticos': Neumaticos,
#       'Marca': Marca,
#       'Precio': Precio})

#   # Agregar la columna "fecha_scrapping" con la fecha actual
# df_neumaticos["fecha_scrapping"] = datetime.date.today()

# # # #exportamos a excel
fecha_scrapping = datetime.date.today()
df_neumaticos.to_excel(f'neumaticos{fecha_scrapping}.xlsx', index=False)
print(f'Datos exportados a neumaticos{fecha_scrapping}.xlsx')


### CODIGO PARA DIVIDR BASES CON PRECIO MAXIMO Y MINIMO

import pandas as pd

# Lista con los nombres de los archivos
archivos = [f'neumaticos{fecha_scrapping}.xlsx']

# Función para procesar cada archivo y guardar en hojas de un único archivo Excel
def procesar_archivo(archivo):
    # Cargar el archivo Excel
    df = pd.read_excel(archivo)

    # Asegurarse de que los datos estén ordenados por 'Neumaticos' y 'Precio'
    df = df.sort_values(by=['Neumaticos', 'Precio'], ascending=[True, False])

    # Dividir los datos en dos: uno con el precio más alto y otro con el más bajo
    df_max = df.groupby('Neumaticos').first().reset_index()
    df_min = df.groupby('Neumaticos').last().reset_index()

    # Nombre base para el archivo Excel (sin extensión)
    nombre_base = archivo.split('.')[0]

    # Crear un objeto ExcelWriter para guardar las dos hojas
    with pd.ExcelWriter(f'{nombre_base}_dividido.xlsx', engine='openpyxl') as writer:
        # Guardar los DataFrames en diferentes hojas del mismo archivo Excel
        df_max.to_excel(writer, sheet_name='Max_Precio', index=False)
        df_min.to_excel(writer, sheet_name='Min_Precio', index=False)

    print(f"Archivo {nombre_base}_dividido.xlsx guardado correctamente.")

# Procesar cada archivo
for archivo in archivos:
    procesar_archivo(archivo)

print("Todos los archivos han sido procesados.")
```

```python
import pandas as pd

# Función para realizar el merge y detectar nuevos neumáticos
def merge_neumaticos(df_base, df_nuevo, fecha_nuevo):
    # Renombrar la columna de precio en el nuevo archivo con la fecha correspondiente
    df_nuevo = df_nuevo.rename(columns={'Precio': fecha_nuevo})

 # Eliminar las columnas duplicadas en el dataframe nuevo
    df_nuevo = df_nuevo.drop(columns=['Marca', 'fecha_scrapping'], errors='ignore')

    # Realizar el merge usando 'outer' para incluir todos los datos de ambos archivos
    merged_df = pd.merge(df_base, df_nuevo, on='Neumaticos', how='outer', indicator=True)

    # Crear dataframe con los neumáticos nuevos (solo presentes en df_nuevo)
    nuevos_neumaticos = merged_df[merged_df['_merge'] == 'right_only'].drop(columns='_merge')

    # Filtrar los datos donde existe coincidencia o solo están en la base existente
    merged_df = merged_df[merged_df['_merge'] != 'right_only'].drop(columns='_merge')

    return merged_df, nuevos_neumaticos

# Cargar el archivo base actualizado o crearlo si no existe
try:
    archivo_base = pd.ExcelFile('neumaticos_actualizados.xlsx')
    max_base = archivo_base.parse('Max_Precio')
    min_base = archivo_base.parse('Min_Precio')
except FileNotFoundError:
    # Si no existe, crear DataFrames vacíos para iniciar la base
    max_base = pd.DataFrame(columns=['Neumaticos'])
    min_base = pd.DataFrame(columns=['Neumaticos'])

# # Cargar el nuevo archivo de neumáticos que deseas agregar y renombrar con la fecha
archivo_nuevo = pd.ExcelFile('neumaticos2024-12-17_dividido.xlsx')
fecha_nuevo = '17/12/2024'  # Cambiar a la fecha correspondiente al nuevo archivo
max_nuevo = archivo_nuevo.parse('Max_Precio').rename(columns={'Precio': fecha_nuevo})
min_nuevo = archivo_nuevo.parse('Min_Precio').rename(columns={'Precio': fecha_nuevo})

## Realizar el merge para ambas hojas con la fecha como encabezado de la columna de precio
max_merged, max_nuevos_neumaticos = merge_neumaticos(max_base, max_nuevo, fecha_nuevo)
min_merged, min_nuevos_neumaticos = merge_neumaticos(min_base, min_nuevo, fecha_nuevo)

# Guardar en el archivo base actualizado
with pd.ExcelWriter('neumaticos_actualizados.xlsx', engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    max_merged.to_excel(writer, sheet_name='Max_Precio', index=False)
    min_merged.to_excel(writer, sheet_name='Min_Precio', index=False)

    # Concatenar y guardar los nuevos neumáticos en una hoja separada
    nuevos_neumaticos = pd.concat([max_nuevos_neumaticos, min_nuevos_neumaticos])
    nuevos_neumaticos.to_excel(writer, sheet_name='nuevos_neumaticos', index=False)
```
